```python
import pandas as pd
import glob
import os 
```

```python
joined_files = os.path.join("data/JC*.csv")
```

```python
joined_list = glob.glob(joined_files)
```

```python
df = pd.concat(map(pd.read_csv,joined_list), ignore_index=True)
df.head()
```

```python
#export cleaned dataset 
df.to_csv('combinedCitiBikes2.csv', index=False)
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
# Convert the start date and end date to have dates and time stamp in different collumns column to datetime format
df['start_date'] = df['started_at'].dt.strftime('%m/%d/%Y')
df['end_date'] = df['ended_at'].dt.strftime('%m/%d/%Y')
df.head()
```

```python
#now convert start time and end time to their own columns 
df['start_time'] = df['started_at'].dt.strftime('%I:%M:%S %p')
df['end_time'] = df['ended_at'].dt.strftime('%I:%M:%S %p')
df.head()
```

```python
#drop start at and end at column
#df = df.drop(['started_at', 'ended_at'], axis=1)
df.head()
```

```python
df.columns 
```

```python
#rearrange columns in df for easier readability 
df = df[['ride_id','rideable_type','member_casual','started_at','start_date','start_time','ended_at',
         'end_date','end_time','start_station_name','start_station_id','end_station_name','end_station_id',
         'start_lat','start_lng','end_lat','end_lng']]
df.head()
```

```python
#export cleaned dataset 
#df.to_csv('combinedCitiBikes.csv', index=False)
```

```python
#find null values
null_df = df[df.isnull().any(axis=1)]
null_df
```

```python
#drop null values
df = df.dropna()
df
```
